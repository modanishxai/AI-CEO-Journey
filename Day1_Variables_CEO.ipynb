```python
# Step 1: I am creating a variable for my name.
ceo_name = "Mo Danish"

# Step 2: I am defining the current year as an integer.
current_year = 2026

# Step 3: I am setting my target year for the vision.
target_year = 2030

# Step 4: I am calculating the remaining years by subtraction.
years_left = target_year - current_year

# Step 5: I am printing the welcome message.
print("Welcome back, " + ceo_name)

# Step 6: I am displaying the years left to achieve the goal.
print("Time remaining for Vision 2030: " + str(years_left) + " years")
```

```python
# Step 1: Integer (Whole Number)
initial_capital = 5000

# Step 2: Float (Decimal Number)
growth_multiplier = 2.5

# Step 3: Math Operation (Multiplication)
# 5000 * 2.5 = 12500.0
projected_valuation = initial_capital * growth_multiplier

# Step 4: String (Text)
currency = "USD"

# Step 5: Printing with Type Casting (Converting number to text)
print("Initial Capital: " + str(initial_capital) + " " + currency)
print("Projected Valuation: " + str(projected_valuation) + " " + currency)

# BONUS: Check the type of data
print("Type of Valuation variable is:")
print(type(projected_valuation))
```

```python
# Step 1: Input - I am asking the user to enter the password.
# The prompt "Enter Password: " will show on the screen.
user_input = input("Enter Password: ")

# Step 2: Logic - I am checking if the input matches the correct password.
# Note: In Python, we use '==' for comparison and ':' to start the block.
if user_input == "Hello CEO":

    # Step 3: Output (True) - If the password is correct.
    print("Access Granted!")
    print("Welcome back, Sir.")

else:

    # Step 4: Output (False) - If the password is wrong.
    print("Access Denied!")
    print("Get out!")
```

```python
print("--- CEO DASHBOARD MENU ---")
print("1. Profit Calculator")
print("2. Age Calculator")

choice = input("Choose an option (1 or 2): ")

# LOGIC SHURU
if choice == "1":
    print("--- Profit Mode ON ---")
    # Ab hum value mangenge (Step by Step)
    sp = int(input("Selling Price batao: "))
    cp = int(input("Cost Price batao: "))

    # Ab Calculation
    profit = sp - cp
    print("Total Profit: " + str(profit))

elif choice == "2":
    print("--- Age Mode ON ---")
    # Ab hum Year mangenge
    current_year = int(input("Current Year: "))
    dob = int(input("Birth Year: "))
    # Ab Calculation
    age = current_year - dob
    print("Your Age is: " + str(age))

else:
    print("Invalid Option! Get Out!")
```
